##### Mathematics within Standardization (Z - Score)
Z-SCORE

Z-Score is one of the most popular methods to standardize data, and can be done by subtracting the mean and dividing by the standard deviation for each value of each feature.

Z-Score standard deviation
Once the standardization is done, all the features will have a mean of zero, a standard deviation of one, and thus, the same scale.

Important Notes
- There exist other standardization methods.
- Remember standardizing the data does change the distribution (shape) of the data but instead the magnitude. This will be clearly evident in our second function which has been designed to visually observe and compare the difference between the data in its original form and the data standardized.

```python
#import square root calculation from math package
from math import sqrt
#import visualization packages
import matplotlib.pyplot as plt
import seaborn as sns
```

### Function 1
###### Perform data stadardization on an arbitrary number of elements.

```python
def ZScore(*args):
    #Create empty list
    list1 = []
    #Initialize sum all *args
    sum_all = 0
    #Accumulate the sum
    for x in args:
        sum_all += x
    #Add all elements 
    for i in args:
        list1.append(i)
    #Find the list length (total count)
    length = len(list1)
    #Calculate the average
    average = sum_all / length
    print('Average',average)
    
    #find the VARIANCE
    list2 = []
    list3 = []
    #calculate the difference between each data point and the average
    for x in args:
        list2.append(x - average)
    #square each value returned from the above calculation
    for i in list2:
        list3.append(i ** 2)
    #Initialize sum all 
    sum_all1 = 0
    #Accumulate the sum of the above (squared) calculation and divdie by number of elements (variance)
    for x in list3:
        sum_all1 += x
    variance = sum_all1 / length
    print('Variance', variance)
    
    #find the standard deviation 
    std = sqrt(variance)
    print('Standard Deviation', std)
    
    #Z - Score Standardization
    standardized1 = []
    standardized2 = []
    for i in args:
        standardized1.append(i - average)
    for x in standardized1:
        standardized2.append(x / std)
    print(standardized2)
```

```python
#Call the function 
ZScore(25,25,45,100)
```

###### Function One Evaluation

The next step is to create a function which we can visually compare the affect of standardizing data in comparison to the orginal date. This will be achieved through a kernel density plot to visualize the distribution of the data.

### Function 2
###### Standardize an arbitrary number of elements and visually compare the dissimilarities via kernel denstiy plot 

```python
def ZScoreVisual(*args):
    #Create empty list
    list1 = []
    #Initialize sum all *args
    sum_all = 0
    #Accumulate the sum
    for x in args:
        sum_all += x
    #Add all elements 
    for i in args:
        list1.append(i)
    print('Original Data', list1)
    #Find the list length (total count)
    length = len(list1)
    #Calculate the average
    average = sum_all / length
    print('Average',average)
    
    #find the VARIANCE
    list2 = []
    list3 = []
    #calculate the difference between each data point and the average
    for x in args:
        list2.append(x - average)
    #square each value returned from the above calculation
    for i in list2:
        list3.append(i ** 2)
    #Initialize sum all 
    sum_all1 = 0
    #Accumulate the sum of the above (squared) calculation and divdie by number of elements (variance)
    for x in list3:
        sum_all1 += x
    variance = sum_all1 / length
    print('Variance :', variance)
    
    #find the standard deviation 
    std = sqrt(variance)
    print('Standard Deviation', std)
    
    #Z - Score Standardization
    standardized1 = []
    standardized2 = []
    for i in args:
        standardized1.append(i - average)
    for x in standardized1:
        standardized2.append(x / std)
    print('Data Standardized', standardized2)
    
    #Comparative Visualizations
    #subplots
    fig,(ax1,ax2) = plt.subplots(1,2)
    #plot size
    fig.set_size_inches(20, 8)
    sns.set_style('darkgrid')
    sns.distplot(args, kde = True, hist = True, color = 'b', ax=ax1)
    sns.distplot(standardized2, kde = True, hist = True, color = 'r', ax=ax2)
    #Adding title
    fig.suptitle('Orignal Data | Z - Score Standardized Data')
    plt.show()
```

```python
#Call Z-Score Visual Function
ZScoreVisual(25,25,45,100)
```

##### Function Two Evaluation

We can clearly observe from the above example visuals on the second function call that the distribution has not changed after the data has been (Z -Score) standardized. This is an important feature to understand regarding this type of standardization. 

Furthermore, we can evidently see that the magnitude of the data has been changed when the original data has been standarized. This is also another important feature to understand when standaridizng data.
