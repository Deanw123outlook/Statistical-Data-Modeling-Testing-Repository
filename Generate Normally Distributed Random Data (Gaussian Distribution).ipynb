### Generate Normally Distributed Random Data (Gaussian Distribution) 

```python
#import required libraries
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Create an array of normaly distributed random data
gauissian = np.random.normal(loc = 0, scale = 1, size = 100)
#Convert np.array to a list
random_normal = gauissian.tolist()
```

```python
#Kernel density plot
sns.distplot(random_normal, kde = True, hist = False, color = 'r')
sns.set_style('darkgrid')
display(plt.show())
```
